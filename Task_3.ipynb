# Model For area & price

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
land= pd.DataFrame({
    'area':[2600,2499,3000,4600,2700,5700,7000],
    'price':[51000,55600,65000,75800,59600,79000,98000]
})
land
```

```python
# spliting data into dependent & independent Variable

x=land[['area']]
x
```

```python
y=land['price']
y
```

```python
plt.title('area vs price')
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(x,y,color='red',marker='o')
plt.show()
```

```python
model= LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.predict([[5000]])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.coef_*5000 + model.intercept_
```

```python
model.predict([[5990]])
```

```python
pred=model.predict(x)
```

```python
land
```

```python
plt.title('area vs price')
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(x,y,color='red',marker='o')
plt.plot(x,pred)
plt.show()
```
